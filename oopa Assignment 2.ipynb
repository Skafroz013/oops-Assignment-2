```python
#16 how do you prevent a class from having multiple instances by using constructors in Python? Provide an example

#We can prevent a class from having multiple instances by using the singleton design pattern
#The singleton pattern ensures that a clss has only one instance, no matter how many times you try to create new instances
class SingletonClass:
    _instance = None  # Class attribute to store the single instance

    def __new__(cls):
        if cls._instance is None:
            cls._instance = super(SingletonClass, cls).__new__(cls)
        return cls._instance

instance1 = SingletonClass()
instance2 = SingletonClass()

print(instance1 is instance2) 
```

```python
#1. What is inheritance in Python? Explain its significance in object-oriented programming.

#inheritance in python is a fundamental concept in object oriented programming (OOP) that allows a new class to inherit the attributes and methods from an exissting class
#The significance of the inheritance is
#Code Reusablity\
#Hierarchy and abstraction
#Polymorphism
#Extensiblity
#Modelling real world relationships
#Example
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        pass  # This method is intended to be overridden in derived classes

class Dog(Animal):
    def speak(self):
        return f"{self.name} says Woof!"

class Cat(Animal):
    def speak(self):
        return f"{self.name} says Meow!"

dog = Dog("Jaanu")
cat = Cat("soonu")

print(dog.speak()) 
print(cat.speak())  
```

```python
#2. Differentiate between single inheritance and multiple inheritance in Python. Provide examples for each.

#Single Inheritance
#Single inheritance is a type of inheritance where a class inherits from only one parent class (superclass or base class).
#in single inheritance, a child class can have only one immediate parent class.
#It promotes a simple and linear class hierarchy.
class Animal:
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"

dog = Dog()
cat = Cat()

print(dog.speak())  # Output: Woof!
print(cat.speak())  # Output: Meow!
```

```python
#Multiple Inheritance
#multiple inheritance is a type of inheritance where a class can inherit from more than one parent class
#it allows for greater flexiblity and ability to combine features from multiple sources
class Animal:
    def speak(self):
        pass

class Mammal:
    def give_birth(self):
        return "Live birth"

class Reptile:
    def lay_eggs(self):
        return "Lay eggs"

class Platypus(Mammal, Reptile):
    def speak(self):
        return "Quack"

platypus = Platypus()

print(platypus.speak())  
print(platypus.give_birth()) 
print(platypus.lay_eggs())  
```

```python
#12. Explain the use of abstract base classes (ABCs) in Python and how they relate to inheritance. Provide an example using the `abc` module

#Abstract base classes in python are a way to define a blueprint for a  class specifying a set of methods that must be implemented by any concrete class
#How they relate to inheritance

#Defining a Blueprint
#Inheritance
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

circle = Circle(5)
rectangle = Rectangle(4, 6)
print(f"Area of the circle with a radius of {circle.radius}: {circle.area()}")
print(f"Area of the rectangle with width {rectangle.width} and height {rectangle.height}: {rectangle.area()}")

```

```python
#13. How can you prevent a child class from modifying certain attributes or methods inherited from a parent class in Python

#In Python, you can control the behavior of inherited attributes or methods in a child class by using encapsulation and access modifiers.
#While Python does not have strict access modifiers like some other programming languages (e.g., private, protected), it follows the principle of "we are all consenting adults here," implying that developers should follow conventions for access control. Here are some ways to prevent a child class from modifying certain attributes or methods inherited from a parent class:

#Use Single Leading Underscore (_):

#Prefix the attribute or method name with a single leading underscore to indicate that it's intended to be a protected member. This is a convention, and it signals to other developers that the attribute or method should not be accessed or modified outside of the class or its subclasses.
class Parent:
    def __init__(self):
        self._protected_attribute = "I am protected"

    def _protected_method(self):
        print("This method is protected")


class Child(Parent):
    def access_protected(self):
        print(self._protected_attribute)  # Okay

    def modify_protected(self):
        self._protected_attribute = "Modified"  # Okay

    def call_protected_method(self):
        self._protected_method()  # Okay


child_instance = Child()
child_instance.access_protected()  # Okay
child_instance.modify_protected()  # Okay
child_instance.call_protected_method()  # Okay
#Name Mangling (Double Leading Underscore):
#Prefix the attribute or method name with a double leading underscore to invoke name mangling. This changes the name of the attribute or method to include the name of the class, making it harder to accidentally override in a subclass.
class Parent:
    def __init__(self):
        self.__mangled_attribute = "I am mangled"

    def __mangled_method(self):
        print("This method is mangled")


class Child(Parent):
    def access_mangled(self):
        print(self._Parent__mangled_attribute)  # Okay

    def modify_mangled(self):
        self._Parent__mangled_attribute = "Modified"  # Okay

    def call_mangled_method(self):
        self._Parent__mangled_method()  # Okay


child_instance = Child()
child_instance.access_mangled()  # Okay
child_instance.modify_mangled()  # Okay
child_instance.call_mangled_method()  # Okay
```

```python
#17. Create a Python class for a library system that encapsulates book information, including titles, authors, and availability status

class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.available = True  

    def check_out(self):
        if self.available:
            self.available = False
            return f"{self.title} by {self.author} has been checked out."
        else:
            return f"{self.title} by {self.author} is already checked out."

    def check_in(self):
        if not self.available:
            self.available = True
            return f"{self.title} by {self.author} has been checked in."
        else:
            return f"{self.title} by {self.author} is already available."

    def is_available(self):
        return self.available


book1 = Book("The Great Gatsby", "F. Scott Fitzgerald")
book2 = Book("To Kill a Mockingbird", "Harper Lee")

print(f"Is '{book1.title}' available? {book1.is_available()}")
print(f"Is '{book2.title}' available? {book2.is_available()}")


print(book1.check_out())
print(f"Is '{book1.title}' available? {book1.is_available()}")

print(book1.check_in())
print(f"Is '{book1.title}' available? {book1.is_available()}")
```

```python
#15  Describe the concept of operator overloading in Python and how it relates to polymorphism. Provide examples using operators like `+` and `*`.

#Operator overloading is a concept in Python that allows you to define and customize the behavior of operators such as +, -, *, /, ==, !=, and more for user-defined objects. 
#This means you can make these operators work with your custom classes and objects in a way that makes sense for your specific application. 
#Operator overloading is closely related to polymorphism, as it enables objects of different types to interact with operators in a consistent and polymorphic manner.
#Example

class ComplexNumber:
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag
    def __str__(self):
        if self.imag < 0:
            return f"{self.real} - {abs(self.imag)}i"
        else:
            return f"{self.real} + {self.imag}i"
    def __add__(self, other):
        real_sum = self.real + other.real
        imag_sum = self.imag + other.imag
        return ComplexNumber(real_sum, imag_sum)
    def __mul__(self, other):
        real_product = (self.real * other.real) - (self.imag * other.imag)
        imag_product = (self.real * other.imag) + (self.imag * other.real)
        return ComplexNumber(real_product, imag_product)
# Create complex numbers
c1 = ComplexNumber(3, 2)
c2 = ComplexNumber(1, -4)
# Demonstrate operator overloading and polymorphism
sum_result = c1 + c2
product_result = c1 * c2
print(f"Sum: {c1} + {c2} = {sum_result}")
print(f"Product: {c1} * {c2} = {product_result}")
```

```python
#6 Create a Python class for a bank account and demonstrate abstraction by hiding the account balance and providing methods to deposit and withdraw funds

from abc import ABC, abstractmethod

class BankAccount(ABC):
    def __init__(self, account_holder, account_number):
        self.account_holder = account_holder
        self.account_number = account_number
        self._balance = 0  # The balance is private and hidden from external access

    @property
    def balance(self):
        return self._balance

    @abstractmethod
    def deposit(self, amount):
        pass

    @abstractmethod
    def withdraw(self, amount):
        pass

class SavingsAccount(BankAccount):
    def __init__(self, account_holder, account_number, interest_rate=0.02):
        super().__init__(account_holder, account_number)
        self.interest_rate = interest_rate

    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            print(f"Deposited ${amount}. New balance: ${self._balance}")
        else:
            print("Invalid deposit amount. Please enter a positive value.")

    def withdraw(self, amount):
        if amount > 0 and amount <= self._balance:
            self._balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self._balance}")
        elif amount <= 0:
            print("Invalid withdrawal amount. Please enter a positive value.")
        else:
            print("Insufficient funds for withdrawal.")

# Example Usage:
savings_account = SavingsAccount(account_holder="John Doe", account_number="123456789")

# Deposit funds
savings_account.deposit(1000)

# Withdraw funds
savings_account.withdraw(500)

# Attempt to withdraw more than the balance
savings_account.withdraw(700)

# Display final balance
print(f"Final balance: ${savings_account.balance}")
```

```python
#13  Create a Python class hierarchy for employees in a company (e.g., manager, developer, designer) and implement abstraction by defining a common `get_salary()` method.

from abc import ABC, abstractmethod

class Employee(ABC):
    def __init__(self, name, employee_id):
        self.name = name
        self.employee_id = employee_id

    @abstractmethod
    def get_salary(self):
        pass

class Manager(Employee):
    def __init__(self, name, employee_id, base_salary, bonus):
        super().__init__(name, employee_id)
        self.base_salary = base_salary
        self.bonus = bonus

    def get_salary(self):
        return self.base_salary + self.bonus

class Developer(Employee):
    def __init__(self, name, employee_id, base_salary, overtime_pay):
        super().__init__(name, employee_id)
        self.base_salary = base_salary
        self.overtime_pay = overtime_pay

    def get_salary(self):
        return self.base_salary + self.overtime_pay

class Designer(Employee):
    def __init__(self, name, employee_id, base_salary, project_bonus):
        super().__init__(name, employee_id)
        self.base_salary = base_salary
        self.project_bonus = project_bonus

    def get_salary(self):
        return self.base_salary + self.project_bonus

# Example Usage:
manager = Manager(name="John Doe", employee_id="M001", base_salary=60000, bonus=15000)
developer = Developer(name="Alice Smith", employee_id="D001", base_salary=50000, overtime_pay=2000)
designer = Designer(name="Bob Johnson", employee_id="DS001", base_salary=55000, project_bonus=5000)

employees = [manager, developer, designer]

for employee in employees:
    print(f"{employee.name} (ID: {employee.employee_id})")
    print(f"Total Salary: ${employee.get_salary()}")
    print()
```

```python
#16 Create a Python class for a computer system, demonstrating abstraction by defining common methods (e.g., `power_on()`, `shutdown()`) in an abstract base class.

from abc import ABC, abstractmethod

class ComputerSystem(ABC):
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
        self.is_powered_on = False

    @abstractmethod
    def power_on(self):
        pass

    @abstractmethod
    def shutdown(self):
        pass

    def restart(self):
        if self.is_powered_on:
            print(f"Restarting {self.brand} {self.model}...")
            self.shutdown()
            self.power_on()
        else:
            print("The computer is not powered on.")

class Desktop(ComputerSystem):
    def power_on(self):
        if not self.is_powered_on:
            print(f"Powering on {self.brand} {self.model} desktop...")
            self.is_powered_on = True
        else:
            print("The desktop is already powered on.")

    def shutdown(self):
        if self.is_powered_on:
            print(f"Shutting down {self.brand} {self.model} desktop...")
            self.is_powered_on = False
        else:
            print("The desktop is already powered off.")

class Laptop(ComputerSystem):
    def power_on(self):
        if not self.is_powered_on:
            print(f"Powering on {self.brand} {self.model} laptop...")
            self.is_powered_on = True
        else:
            print("The laptop is already powered on.")

    def shutdown(self):
        if self.is_powered_on:
            print(f"Shutting down {self.brand} {self.model} laptop...")
            self.is_powered_on = False
        else:
            print("The laptop is already powered off.")

# Example Usage:
desktop = Desktop(brand="Dell", model="Inspiron")
laptop = Laptop(brand="HP", model="Envy")

desktop.power_on()
desktop.shutdown()
desktop.restart()

print()

laptop.power_on()
laptop.shutdown()
laptop.restart()
```

```python
#19  Create a Python class for a library system, implementing abstraction by defining common methods (e.g., `add_book()`, `borrow_book()`) in an abstract base class.

from abc import ABC, abstractmethod

class LibrarySystem(ABC):
    def __init__(self):
        self.books = []

    @abstractmethod
    def add_book(self, book):
        pass

    @abstractmethod
    def borrow_book(self, book_title, user):
        pass

    @abstractmethod
    def return_book(self, book_title):
        pass

    def display_books(self):
        if not self.books:
            print("The library is empty.")
        else:
            print("Available Books:")
            for book in self.books:
                print(f"- {book.title} by {book.author}")

class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_borrowed = False

class PublicLibrary(LibrarySystem):
    def add_book(self, book):
        self.books.append(book)
        print(f"Added {book.title} by {book.author} to the library.")

    def borrow_book(self, book_title, user):
        for book in self.books:
            if book.title == book_title and not book.is_borrowed:
                book.is_borrowed = True
                print(f"{user} borrowed {book.title} by {book.author}.")
                return
        print(f"Book '{book_title}' is either not available or already borrowed.")

    def return_book(self, book_title):
        for book in self.books:
            if book.title == book_title and book.is_borrowed:
                book.is_borrowed = False
                print(f"Returned {book.title} by {book.author}.")
                return
        print(f"Book '{book_title}' is not valid or not currently borrowed.")

# Example Usage:
library = PublicLibrary()

book1 = Book(title="Introduction to Python", author="John Doe")
book2 = Book(title="Data Structures in Java", author="Jane Smith")

library.add_book(book1)
library.add_book(book2)

library.display_books()

library.borrow_book("Introduction to Python", user="Alice")
library.borrow_book("Data Structures in Java", user="Bob")

library.display_books()

library.return_book("Introduction to Python")
library.return_book("Non-existent Book")

library.display_books()
```

```python
#12  Create a Python class for a university course, using composition to represent students, instructors, and course materials.

class Student:
    def __init__(self, student_id, name):
        self.student_id = student_id
        self.name = name

    def enroll(self, course):
        print(f"{self.name} enrolled in the course: {course.title}")

    def submit_assignment(self, assignment):
        print(f"{self.name} submitted assignment: {assignment}")


class Instructor:
    def __init__(self, instructor_id, name):
        self.instructor_id = instructor_id
        self.name = name

    def teach(self, course):
        print(f"{self.name} is teaching the course: {course.title}")

    def grade_assignment(self, assignment, student):
        print(f"{self.name} graded assignment '{assignment}' submitted by {student.name}")


class Course:
    def __init__(self, title, code, instructor, students=None, materials=None):
        self.title = title
        self.code = code
        self.instructor = instructor  # Composition: Course has an Instructor
        self.students = students if students is not None else []  # Composition: Course has Students
        self.materials = materials if materials is not None else []  # Composition: Course has Materials

    def add_student(self, student):
        self.students.append(student)

    def add_material(self, material):
        self.materials.append(material)

    def display_info(self):
        print(f"Course: {self.title} ({self.code})")
        print(f"Instructor: {self.instructor.name}")

        print("Students:")
        for student in self.students:
            print(f"- {student.name}")

        print("Course Materials:")
        for material in self.materials:
            print(f"- {material}")


# Example usage:
if __name__ == "__main__":
    # Creating students
    student1 = Student(student_id="S001", name="Alice")
    student2 = Student(student_id="S002", name="Bob")

    # Creating an instructor
    instructor = Instructor(instructor_id="I001", name="Dr. Smith")

    # Creating a course with students and an instructor
    course = Course(title="Introduction to Python", code="CS101", instructor=instructor, students=[student1, student2])

    # Enrolling students in the course
    for student in course.students:
        student.enroll(course)

    # Adding course materials
    course.add_material("Syllabus")
    course.add_material("Lecture Slides")

    # Displaying course information
    course.display_info()

    # The instructor teaches the course
    instructor.teach(course)

    # Students submit assignments
    for student in course.students:
        student.submit_assignment("Assignment 1")

    # The instructor grades assignments
    for student in course.students:
        instructor.grade_assignment("Assignment 1", student)
```

```python
#14  Create a Python class hierarchy for a restaurant system, using composition to represent menus, dishes, and ingredients.

class Ingredient:
    def __init__(self, name, quantity):
        self.name = name
        self.quantity = quantity

    def use(self):
        print(f"Using {self.quantity} {self.name}")


class Dish:
    def __init__(self, name, ingredients):
        self.name = name
        self.ingredients = ingredients  # Composition: Dish has Ingredients

    def cook(self):
        print(f"Cooking {self.name}")
        for ingredient in self.ingredients:
            ingredient.use()


class Menu:
    def __init__(self, name, dishes):
        self.name = name
        self.dishes = dishes  # Composition: Menu has Dishes

    def display_menu(self):
        print(f"Menu: {self.name}")
        for dish in self.dishes:
            print(f"- {dish.name}")


# Example usage:
if __name__ == "__main__":
    # Creating ingredients
    ingredient1 = Ingredient(name="Tomato", quantity=3)
    ingredient2 = Ingredient(name="Cheese", quantity=200)
    ingredient3 = Ingredient(name="Flour", quantity=500)

    # Creating dishes with ingredients
    pizza_ingredients = [ingredient1, ingredient2, ingredient3]
    pizza = Dish(name="Pizza", ingredients=pizza_ingredients)

    pasta_ingredients = [ingredient1, ingredient3]
    pasta = Dish(name="Pasta", ingredients=pasta_ingredients)

    # Creating a menu with dishes
    italian_menu = Menu(name="Italian Cuisine", dishes=[pizza, pasta])

    # Displaying the menu
    italian_menu.display_menu()

    # Cooking a dish from the menu
    selected_dish = italian_menu.dishes[0]
    selected_dish.cook()
```

```python
#16  Create a Python class for a computer game character, using composition to represent attributes like weapons, armor, and inventory.

class Weapon:
    def __init__(self, name, damage):
        self.name = name
        self.damage = damage

    def attack(self):
        print(f"Attacking with {self.name}, dealing {self.damage} damage")


class Armor:
    def __init__(self, name, defense):
        self.name = name
        self.defense = defense

    def defend(self):
        print(f"Wearing {self.name} for {self.defense} defense")


class Inventory:
    def __init__(self):
        self.items = []

    def add_item(self, item):
        self.items.append(item)
        print(f"Added {item.name} to the inventory")

    def view_inventory(self):
        print("Inventory:")
        for item in self.items:
            print(f"- {item.name}")


class GameCharacter:
    def __init__(self, name, health, weapon=None, armor=None, inventory=None):
        self.name = name
        self.health = health
        self.weapon = weapon  # Composition: GameCharacter has a Weapon
        self.armor = armor  # Composition: GameCharacter has an Armor
        self.inventory = inventory if inventory is not None else Inventory()  # Composition: GameCharacter has an Inventory

    def equip_weapon(self, weapon):
        self.weapon = weapon
        print(f"{self.name} equipped with {weapon.name}")

    def equip_armor(self, armor):
        self.armor = armor
        print(f"{self.name} equipped with {armor.name}")

    def take_damage(self, damage):
        effective_damage = max(0, damage - (self.armor.defense if self.armor else 0))
        self.health -= effective_damage
        print(f"{self.name} took {effective_damage} damage. Health: {self.health}")

    def use_item(self, item):
        print(f"{self.name} is using {item.name}")
        # Add logic for using items, if needed
        self.inventory.items.remove(item)

    def display_status(self):
        print(f"{self.name}'s Status:")
        print(f"Health: {self.health}")
        if self.weapon:
            print(f"Weapon: {self.weapon.name} (Damage: {self.weapon.damage})")
        if self.armor:
            print(f"Armor: {self.armor.name} (Defense: {self.armor.defense})")
        self.inventory.view_inventory()


# Example usage:
if __name__ == "__main__":
    # Creating weapons and armor
    sword = Weapon(name="Sword", damage=20)
    bow = Weapon(name="Bow", damage=15)
    helmet = Armor(name="Helmet", defense=5)
    shield = Armor(name="Shield", defense=10)

    # Creating a character with weapons, armor, and inventory
    player = GameCharacter(name="Hero", health=100, weapon=sword, armor=helmet)

    # Displaying the initial status
    player.display_status()

    # Player equips a different weapon and armor
    player.equip_weapon(bow)
    player.equip_armor(shield)

    # Displaying the updated status
    player.display_status()

    # Player takes damage
    player.take_damage(25)

    # Adding items to the inventory
    player.inventory.add_item(Weapon(name="Dagger", damage=10))
    player.inventory.add_item(Armor(name="Cape", defense=3))

    # Using an item from the inventory
    player.use_item(player.inventory.items[0])

    # Displaying the final status
    player.display_status()
```

```python
#17 Describe the concept of "aggregation" in composition and how it differs from simple composition

#In object-oriented programming, "aggregation" is a form of composition where one class contains another class as a part, but the contained class can exist independently of the containing class. Aggregation represents a "has-a" relationship, indicating that one class has a collection of objects or components, but the lifecycle of the contained objects is not tied to the lifecycle of the containing object.

#Here are key characteristics and differences between simple composition and aggregation:

#Independence of Contained Objects:

#In aggregation, the contained objects can exist independently of the containing object. They can be created and destroyed without affecting the existence of the containing object.
#In simple composition, the contained objects are typically created and managed by the containing object, and their lifecycle is closely tied to the lifecycle of the container.
#Multiplicity:

#Aggregation often involves a "many" relationship, where one object contains a collection of other objects. The contained objects may have a multiplicity greater than one.
#Simple composition may involve a "one" relationship, where one object contains another object, and the multiplicity is often limited to one.
#Example:

#Aggregation might be used to represent a university and its departments. A university contains multiple departments, and each department can exist independently of the university. If the university is closed, the departments can still exist.
#Simple composition might be used to represent a car and its engine. A car contains an engine, and the engine's existence is closely tied to the car. If the car is scrapped, the engine is typically discarded as well.
#Code Representation:

#In code, aggregation is often represented by having a reference to another class as an attribute of the containing class. This reference can be created and set externally, indicating that the contained object is independent.
#In simple composition, the contained object is often created and managed directly within the containing class.
```

```python
#18  Create a Python class for a house, using composition to represent rooms, furniture, and appliances.

class Furniture:
    def __init__(self, name):
        self.name = name

    def arrange(self):
        print(f"Arranging {self.name}")


class Appliance:
    def __init__(self, name):
        self.name = name

    def turn_on(self):
        print(f"{self.name} is turned on")

    def turn_off(self):
        print(f"{self.name} is turned off")


class Room:
    def __init__(self, name, furniture=None, appliances=None):
        self.name = name
        self.furniture = furniture if furniture is not None else []  # Composition: Room has Furniture
        self.appliances = appliances if appliances is not None else []  # Composition: Room has Appliances

    def add_furniture(self, furniture):
        self.furniture.append(furniture)

    def add_appliance(self, appliance):
        self.appliances.append(appliance)

    def describe(self):
        print(f"Room: {self.name}")
        print("Furniture:")
        for item in self.furniture:
            print(f"- {item.name}")
        print("Appliances:")
        for appliance in self.appliances:
            print(f"- {appliance.name}")


class House:
    def __init__(self, rooms):
        self.rooms = rooms  # Composition: House has Rooms

    def describe_house(self):
        print("House Description:")
        for room in self.rooms:
            room.describe()


# Example usage:
if __name__ == "__main__":
    # Creating furniture and appliances
    sofa = Furniture(name="Sofa")
    dining_table = Furniture(name="Dining Table")
    fridge = Appliance(name="Refrigerator")
    oven = Appliance(name="Oven")

    # Creating rooms with furniture and appliances
    living_room = Room(name="Living Room", furniture=[sofa], appliances=[fridge])
    kitchen = Room(name="Kitchen", furniture=[dining_table], appliances=[oven])

    # Creating a house with rooms
    my_house = House(rooms=[living_room, kitchen])

    # Describing the house
    my_house.describe_house()
```
